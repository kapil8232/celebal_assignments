# Step 4: Handle Missing Values

# 4.1 Identify Missing Values

```python
missing_values = df.isnull().sum()
missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing_Count': missing_values.values
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count',ascending = False)
print('Printing Missing Values:')
missing_df
```

 # 4.2 Handling Missing Values

```python
# Fill missing values in numerical columns with median
for col in df.select_dtypes(include=['int64', 'float64']):
    df[col].fillna(df[col].median(), inplace=True)

# Fill missing values in categorical columns with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

# Check remaining missing values
print(df.isnull().sum().sum())
```

# Step 5: Filtering

```python
# Select specific columns
important_cols = ['product_id', 'title', 'category', 'initial_price', 'discount', 'rating', 'ratings_count']
df_selected = df[important_cols].copy()

print(df_selected.head(5))
```

```python
# Filter products with high ratings
high_rated = df[df['rating'] >= 4.0]
print(f'Products with rating >= 4.0: {len(high_rated)} out of {len(df)}')

print('\nSample of high-rated products:')
print(high_rated[['title', 'rating', 'ratings_count']].head(10))
```

```python
# Filter products with significant discounts (>50%)
high_discount = df[df['discount'] > 50]
print(f'Products with discount > 50%: {len(high_discount)} out of {len(df)}')

print('\nSample of high-discount products:')
print(high_discount[['title', 'initial_price', 'discount']].head(10))
```

# Step 6: Remove Duplicates

```python
# Check for duplicates
print(f'Total rows before removing duplicates: {len(df)}')
print(f'Duplicate rows based on product_id: {df.duplicated(subset=["product_id"]).sum()}')

# Remove duplicates
df = df.drop_duplicates(subset=['product_id'], keep='first')
print(f'Total rows after removing duplicates: {len(df)}')
```

# Step 7: Create Derived Columns

```python
# Create a derived column: discount_amount (savings in rupees)
df['discount_amount'] = (df['initial_price'] * df['discount']) / 100

# Create a column: final_price (calculated from initial_price and discount)
df['final_price_calculated'] = df['initial_price'] - df['discount_amount']

# Create a quality score based on rating and number of ratings
df['quality_score'] = (df['rating'] * df['ratings_count']) / df['ratings_count'].max()

print('\nSample of new columns:')
print(df[['title', 'initial_price', 'discount', 'discount_amount', 'final_price_calculated']].head(10))
```

# Step 8: Final Data Validation

```python
# Validate data quality
print('Final Data Validation:')
print(f'Total rows: {len(df)}')
print(f'Total columns: {len(df.columns)}')
print(f'Missing values: {df.isnull().sum().sum()}')
print(f'Duplicate rows: {df.duplicated().sum()}')
print(f'\nData quality: PASSED ✓')
```

# Step 9: Save Cleaned Dataset

```python
# Save cleaned dataset
output_file = 'Combined_dataset_CLEANED.csv'
df.to_csv(output_file, index=False)
```
