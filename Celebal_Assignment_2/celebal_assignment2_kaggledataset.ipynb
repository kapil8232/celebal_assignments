## STEP 1: Import Libraries 

```python
import sqlite3
import pandas as pd
print("Libraries imported")
```

## Step 2: Load the data into Sql Database

```python
#Load Dataset
df = pd.read_csv("Superstores.csv", encoding="latin1")
print(f"Shape is:{df.shape}")
print(f"Rows are:{df.shape[0]}")
print(f"Columns are:{df.shape[1]}")
```

##  Make connection of sqlite

```python
# Create SQLite database
conn = sqlite3.connect("superstore.db")
cursor=conn.cursor()
```

## Load data into SQL table

```python
# Creating table superstore
df.to_sql("superstore", conn, if_exists="replace", index=False)
print("Dataset Loaded Successfully")
```

## STEP 3: Explore Table Schema and Sample Data

```python
# query execution function
def run_query(query):
    return pd.read_sql(query,conn)
```

```python
q1= """
PRAGMA table_info(superstore);
"""
df_q1=run_query(q1)
df_q1
```

```python
q2="""
select *
from superstore
LIMIT 5
"""
df_q2=run_query(q2)
df_q2
```

```python
q3= """
SELECT COUNT(*) AS total_rows
FROM superstore;
"""
df_q3=run_query(q3)
print("Total Rows:")
df_q3
```

## STEP 4: Filtering Data

```python
q4="""
select *
from superstore
where Region = 'West'
"""
print("Data in Region West:")
print("""Insight:Filtering data by region helps analyze sales performance in a specific geographical area.""")
df_q4=run_query(q4)
df_q4
```

```python
q5="""
select *
from superstore
where Category = 'Technology';
"""
print("Data in Technology category:\n")
print("""Insight:Filtering on category enables analysis of products such as Furniture, Office Supplies, or Technology.""")
df_q5=run_query(q5)
df_q5
```

```python
q5="""
select *
from superstore
where Sales > 2000
"""
print("Data in when sales >2000:\n")
print("""Insight:High-value sales transactions can help to understand major revenue contributors.""")
df_q5=run_query(q5)
df_q5
```

```python
q6="""
select *
from superstore
where City ='Los Angeles'
"""
print("Data in Los Angeles:\n")

df_q6=run_query(q6)
df_q6
```

```python
q7="""
select *
from superstore
where "Order Date" like '%2017';
"""
print("Data in year 2017:\n")
print("""Insight:Date-based filtering help to analysis of sales within a specific  time period""")
df_q7=run_query(q6)
df_q6
```

## Step 5:GROUP BY  And Aggregations Functions

```python
q8="""
select Region,
       SUM(Sales) AS Total_Sales
from superstore
GROUP BY Region
"""
print("Total Sales by Region")
print("""Insight:This query shows which regions contribute the most revenue and helps identify strong and weak region.""")
df_q8=run_query(q8)
df_q8.style.hide(axis="index")
```

```python
q9="""
select Category,
       SUM(Quantity) AS Total_Quantity
from superstore
GROUP BY Category
"""
print("Total Quantity by Category")
print("""Insight:This query indicates which product categories have the highest customer demand """)
df_q9=run_query(q9)
df_q9.style.hide(axis="index")
```

```python
q10="""
select Category,
       AVG(Profit) AS Avg_Profit
from superstore
GROUP BY Category
"""
print("Average Profit by Category")
print("""Insight:Comparing average profits across categories helps to determine which categories are most profitable""")
df_q10=run_query(q10)
df_q10.style.hide(axis="index")
```

## Step 5 :Sort and Limit clauses

```python
q11="""
select "Product Name",
       SUM(Sales) AS Total_Sales
from superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10
"""
print("Top 10 Products by Sales")
print("""Insight:This query help to determine highest-selling products and revenue-generating items.""")
df_q11=run_query(q11)
df_q11.style.hide(axis="index")
```

```python
q12="""
SELECT Category,
       ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore
GROUP BY Category
ORDER BY Total_Profit DESC
"""
print("Top Categories by Profit")
print("""Insight:Profit-based ranking reveals which product categories contribute most to business profitabilit""")
df_q12=run_query(q12)
df_q12.style.hide(axis="index")
```

## step 6. Business Use Cases

```python
q13="""
select SUBSTR("Order Date",-4) AS Year,
       SUBSTR("Order Date",1,2) AS Month,
       ROUND(SUM(Sales),2) AS Total_Sales
from superstore
GROUP BY Year, Month
ORDER BY Year, Month
"""
print("Monthly Sales Trend")
print("""Insight:Monthly sales trends reveal seasonality patterns, growth periods, and months with weaker performance""")
df_q13=run_query(q13)
df_q13.style.hide(axis="index")
```

```python
q14="""
select State,
       SUM(Profit) AS Total_Profit
from superstore
GROUP BY State
ORDER BY Total_Profit DESC
LIMIT 10
"""
print("Most Profitable States:")
df_q14=run_query(q14)
df_q14.style.hide(axis="index")
```

```python
q15="""
select "Order ID",
       COUNT(*) AS Duplicate_Count
from superstore
GROUP BY "Order ID"
HAVING Duplicate_Count > 10
"""
print("Find Duplicate Orders:")
print("""Insight:Duplicate Order IDs  represent multiple products within the same order""")
df_q15=run_query(q15)
df_q15.style.hide(axis="index")
```

## Step 7. Validate the result data

```python
q16="""
select
COUNT(*) -
COUNT("Order ID") AS Missing_OrderID,
COUNT(*) -
COUNT(Sales) AS Missing_Sales
from superstore
"""
print("Null Values:")
print("""Insight:Checking for missing values ensures data completeness and improves the reliability of final result""")
df_q16=run_query(q16)
df_q16.style.hide(axis="index")
```

```python
q17="""
select COUNT(DISTINCT "Customer ID") AS Total_Customers
from superstore
"""
print("Distinct customers:")
print("""Insight:The number of unique customers provides an overview of the customer base and market reach""")
df_q17=run_query(q17)
df_q17.style.hide(axis="index")
```
